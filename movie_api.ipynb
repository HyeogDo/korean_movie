### 필요 모듈 불러오기

```python
import numpy as np
import requests
import json
```

### 공통 코드 조회하기

#### 국적 코드 조회하기

```python
# key 값, comCode : 요청 코드 값(국적코드 : "2204", 영화유형코드 : "2201")
# git에 올릴 때 꼭 빼고 올리기!!!!!!!!!!!
key = ""
comCode = "2204"
uri = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/code/searchCodeList.json?key={key}&comCode={comCode}"

# API의 URL 구성하기
url = uri.format(key = key, comCode = comCode)

# API에 요청을 보내 데이터 추출하기
r = requests.get(url)

# 결과를 Json형식으로 변환
data = json.loads(r.text)
```

```python
repNationCd = data["codes"][0]["fullCd"]
repNationCd
```

#### 영화 유형 코드 조회하기

```python
comCode = "2201"
uri = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/code/searchCodeList.json?key={key}&comCode={comCode}"

# API의 URL 구성하기
url = uri.format(key = key, comCode = comCode)

# API에 요청을 보내 데이터 추출하기
r = requests.get(url)

# 결과를 Json형식으로 변환
data = json.loads(r.text)
```

```python
# 장편을 대상으로 추출
data
```

```python
movieTypeCd = data["codes"][0]["fullCd"]
movieTypeCd
```

### 요청 변수 키값들 설정하기

```python
# key : key값, openStartDt : 시작 개봉연도, openEndDt : 마지막 개봉연도, itemPerPage : 한 페이지당 결과 Row 수, curPage : 현재 페이지
key = "fd5d23725f8b05cc91cbca3848ac7437"
openStartDt = "2000"
openEndDt = "2020"
itemPerPage = 100
curPage = 1
```

```python
uri = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key={key}&openStartDt={openStartDt}&openEndDt={openEndDt}&itemPerPage={itemPerPage}" + \
    "&curPage={curPage}&repNationCd={repNationCd}&movieTypeCd={movieTypeCd}"
```

### 영화 목록 불러오기

```python
# API의 URL 구성하기
url = uri.format(key = key, openStartDt = openStartDt, openEndDt = openEndDt, itemPerPage = itemPerPage, 
                    curPage = curPage, repNationCd = repNationCd, movieTypeCd = movieTypeCd)

# API에 요청을 보내 데이터 추출하기
r = requests.get(url)

# 결과를 Json형식으로 변환
data = json.loads(r.text)
```
